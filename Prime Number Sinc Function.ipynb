# Prime Number Sinc Function

https://docs.sympy.org/latest/index.html

```python
import sympy
from sympy import symbols
from sympy import Product
from sympy.utilities.lambdify import lambdify, implemented_function
```

```python
x, n, N = symbols('x n N')
pi = sympy.pi
```

```python
sympy.sin(x)
```

```python
def sn(x,_n):
    return sympy.sin(pi * (x - _n) /_n) / (pi * (x - _n) / _n)
```

```python
sympy.simplify(sn(x,2))
```

```python
from sympy.abc import n, m
def SN(x, N):
    return Product(sn(x, n), (n, 1, N))
```

```python
SN_num = lambdify([x, N], SN(x, N))
```

```python
SN(x,3)
```

### Plotting
